# Q. 1) Instruction ( For Data Cleaning ) - Remove the column that only contains missing values.

```python
drop_missing = Data.drop(missing_column, axis=1)
```

# Q. 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ? 

```python
speeding_df = Data[Data['violation'] == 'Speeding'] ## Filter dataset for speeding incidents

gender_counts = speeding_df['driver_gender'].value_counts() # Count occurrences by gender

if gender_counts['M'] > gender_counts['F']:
    
    print('Men were stopped more often for speeding.')
    
elif gender_counts['M'] < gender_counts['F']:
    
    print('Women were stopped more often for speeding.')
    
else:
    
    print('Men and women were stopped for speeding in equal numbers.')

```

# Q. 3) Question ( Groupby ) - Does gender affect who gets searched during a stop ?

```python
gender_groups = Data.groupby('driver_gender')
```

```python
search_counts = gender_groups['search_conducted'].value_counts(normalize=True)
```

```python
if search_counts['M', True] > search_counts['F', True]:
    
    print("Gender does affect who gets searched during a stop. Men are more likely to be searched.")
    
elif search_counts['M', True] < search_counts['F', True]:
    
    print("Gender does affect who gets searched during a stop. Women are more likely to be searched.")
    
else:
    
    print("Gender does not significantly affect who gets searched during a stop.")
```
